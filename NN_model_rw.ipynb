# Import

```python
import os, sys, math, time, winsound, random, h5py
import numpy as np, pandas as pd, seaborn as sns
import tensorflow as tf
# from matplotlib import pyplot as plt
from tensorflow import keras
from keras.models import model_from_json
from tensorflow.keras import layers
from scipy import stats
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
print(tf.__version__)
```

# 学習用データ作成

## 作成

```python
data_length=1000
space_width=5
```

```python
x=pd.Series([random.uniform(-space_width, space_width) for i in range(data_length)], name="x")
y=pd.Series([random.uniform(-space_width, space_width) for i in range(data_length)], name="y")
```

```python
data=pd.concat([x, y], axis=1)
data.loc[data.x**2+data.y**2<space_width**2/4, "class"]=0
data.loc[data.x**2+data.y**2>=space_width**2/4, "class"]=1
```

```python
x=pd.Series([random.uniform(-space_width, space_width) for i in range(data_length)], name="x")
y=pd.Series([random.uniform(-space_width, space_width) for i in range(data_length)], name="y")
testdata=pd.concat([x, y], axis=1)
```

## 特徴量追加

```python
data["X1"]=data.x.copy()
data["X2"]=data.y.copy()
data.loc[data.x>=space_width/2, "X1"]=space_width/2
data.loc[data.x<=-space_width/2, "X1"]=-space_width/2
data.loc[data.y>=space_width/2, "X2"]=space_width/2
data.loc[data.y<=-space_width/2, "X2"]=-space_width/2
```

```python
testdata["X1"]=testdata.x.copy()
testdata["X2"]=testdata.y.copy()
testdata.loc[testdata.x>=space_width/2, "X1"]=space_width/2
testdata.loc[testdata.x<=-space_width/2, "X1"]=-space_width/2
testdata.loc[testdata.y>=space_width/2, "X2"]=space_width/2
testdata.loc[testdata.y<=-space_width/2, "X2"]=-space_width/2
```

```python
sns.jointplot(x='x', y='y', data=data, hue="class")
```

```python
# data=data.drop(columns=["X1", "X2"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop(columns=["class"]), data["class"], test_size=0.2, random_state=42)
# x_train=(x_train+2)/4
# y_train=y_train
# x_test=(x_test+2)/4
# y_test=y_test
```

## RUS

```python
rus = RandomUnderSampler(random_state=0)
x_train, y_train = rus.fit_resample(x_train, y_train)
x_test, y_test = rus.fit_resample(x_test, y_test)
```

# モデル作成

```python
model = keras.models.Sequential([
#   layers.Dense(5, activation='tanh'),
  layers.Dense(3, activation='tanh'),
  layers.Dense(4, activation='tanh'),
  layers.Dense(2)
])
predictions = model(np.array(x_train)[:1]).numpy()
loss_fn = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, verbose=0)
model.evaluate(x_test, y_test)
# test_result=pd.concat([x_test.reset_index(drop=True), y_test.reset_index(drop=True), pd.Series(np.argmax(model.predict(x_test), axis = 1), name="predict")], axis=1)
# sns.jointplot(x='x', y='y', data=test_result, hue="predict", marker="o")
test_result_2=pd.concat([testdata, pd.Series(np.argmax(model.predict(testdata, verbose=0), axis = 1), name="predict")], axis=1)
sns.jointplot(x='x', y='y', data=test_result_2, hue="predict", marker="o")
```

```python
model.save("model.hdf5")
model.save_weights("weight.hdf5")

with open("model.json", 'w') as f:
    f.write(model.to_json())
```

# モデル読み込み&編集

```python
name_list=[]
with h5py.File("weight.hdf5",'r+') as f, h5py.File("weight2.hdf5",'w') as f2:
    f.visit(printall)
    for name in name_list:
        if isinstance(f[name], h5py.Group):
            f2.create_group(name)
#             print("Group created:",name)
        elif isinstance(f[name], h5py.Dataset):        
            f2.create_dataset(name, data=f[name][()])
#             print("Dataset created:", name, f2[name][()])
```
